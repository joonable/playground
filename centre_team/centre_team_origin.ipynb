```python
for cand1 in pr_queue:
    while True:
        cand2 = random.choice(queue)
        
        if cand1 == cand2 or cand2 in pr_queue: 
            continue
        
        df.loc[cand1]
        df.loc[cand2]

        if df.loc[cand1, 'department'] == df.loc[cand2, 'department']: 
            continue
        
        if df.loc[cand1, 'sex'] == df.loc[cand2, 'sex']: 
            continue

        team_list.append((df.loc[cand1, 'name'], df.loc[cand2, 'name']))
        
        queue.remove(cand1)
        queue.remove(cand2)
        break
```

```python
df_team['mem1_name'] = [tup[0] for tup in team_list]
df_team['mem1_depart'] = [df.loc[df.name == tup[0]].department.values[0] for tup in team_list]
df_team['mem1_sex'] = [df.loc[df.name == tup[0]].sex.values[0] for tup in team_list]


df_team['mem2_name'] = [tup[1] for tup in team_list]
df_team['mem2_depart'] = [df.loc[df.name == tup[1]].department.values[0] for tup in team_list]
df_team['mem2_sex'] = [df.loc[df.name == tup[1]].sex.values[0] for tup in team_list]
```
