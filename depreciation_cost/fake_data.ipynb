```python
import pandas as pd
import numpy as np
import random
```

```python
df_2018 = pd.read_excel('./depreciation_cost_original.xlsx', sheet_name=0)
df_2019 = pd.read_excel('./depreciation_cost_original.xlsx', sheet_name=1)
df = df_2019
```

```python
df.columns
```

```python
df.columns = ['id', 'sub_id', 'asset_class_id', 'asset_class_name', 'asset_name', 'asset_details', 'cate_id', 'cate_name', 
              'intro_year', 'intro_date', 'val', 'depart_id', 'depart_name', 'business_area', 'cost_depart_id', 
              'cost_depart_name', 'years', 'serial_no', 'maintenance_cost', 'maintenance_ratio', 
              'warranty_period', 'original_price']
```

```python
df.info()
```

```python
df = df[['id', 'sub_id', 'asset_class_id', 'asset_class_name', 'asset_name',  'cate_id', 'cate_name', 
          'intro_year', 'intro_date', 'val', 'depart_id', 'depart_name', 'business_area', 'cost_depart_id', 
          'cost_depart_name', 'years', 'original_price']]
```

```python
df = df.drop_duplicates(subset=df.columns[0])
df.iloc[:, 0] = [i for i in range(len(df))]
```

```python
df.head()
```

```python
def randomise_id_name(df, colname):
    num = len(df[colname+'_id'].unique())
    df[colname+'_id'] = np.random.choice(num, size=len(df))
    name_list = df[colname+'_name'].unique()
    df[colname+'_name'] = df.apply(lambda row: name_list[row[colname+'_id']], axis=1)
    return df    

df = randomise_id_name(df, 'asset_class')
df = randomise_id_name(df, 'cate')
df = randomise_id_name(df, 'depart')
df = randomise_id_name(df, 'cost_depart')
```

```python
df.head()
```

```python
def fake_date():
    year = str(np.random.randint(low=2014, high=2019, size=1)[0])
    month = str(np.random.randint(low=1, high=13, size=1)[0]).zfill(2)
    day = str(np.random.randint(low=1, high=29, size=1)[0]).zfill(2)
    return year + '.' + month + '.' + day

df.intro_date = df.intro_date.apply(lambda x: fake_date())
df.intro_year = df.intro_date.apply(lambda _datetime: _datetime[0:4])
```

```python
df.head()
```

```python
def fake_price():
    return 10000 * np.random.randint(low=1, high=100, size=1)[0] + np.random.randint(low=1, high=10000, size=1)[0]

df.original_price = df.original_price.apply(lambda x: fake_price())
```

```python
df.head()
```

```python
asset_name_list = df.asset_name.unique()
num = len(asset_name_list)

df['asset_name'] = df.asset_name.apply(lambda x: asset_name_list[np.random.randint(0, num, 1)[0]])
```
