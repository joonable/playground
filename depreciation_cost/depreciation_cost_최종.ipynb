입력받은 row에 대해서 intro_date를 년(4자리),월(2자리)로 슬라이싱하는 함수를 선언한다.
함수에서의 return값은 코드가 호출되는 reset_intro_date(row) 곳에서 반환된다.
apply 함수에서 axis를 1로 두는 이유 row별로 하나씩 내려가면서 dataframe을 생성하기 때문이다.

```python
def reset_intro_date(row):
    intro_date = row.intro_date
    row['intro_year'] = int(intro_date[:4])
    row['intro_month'] = int(intro_date[5:7]) 
    # row['intro_day'] = int(intro_date[-2:])
    return row


df_depre = df_depre.apply(lambda row: reset_intro_date(row), axis=1)
```

```python
df_depre.head()
```

만료되는 년도는 현재 intro_year에 4년을 더하면 그 제품 혹은 소프트웨어의 감가삼각비는 0가 된다.

```python
df_depre['expired_year'] = df_depre['intro_year'] + 4 
```

48로 나누는 이유는 월당 만원씩 줄기 때문에 4년 뒤는 0이 되므로 48로 나누는 것 이다.

```python
df_depre['monthly_cost'] = df_depre['original_price'] / 48
```

```python
제품별 depre_month : 제품별 감각상각비를 구하는데 해당되는 개월수
제품별 expired_year : 감각상각비가 0원이 되는 년도 수
std_year : 감각 상각비를 추정하고 싶은 년도

조건 1 -  std_year vs. expired_year 
std_year < expired_year  :  기준년 이전에 감가상각비가 0원이 되는 경우
감가상각비가 이미 소진 되었으므로 0개월 부여

std_year > expired_year  :  기준년 이후에 감가상각비가 0원이 되는 경우
1년을 기준으로 감가상각비를 구하므로 12개월을 부여

# std_year == expired_year  :  기준년에 감가상각비가 0원이 되는 경우
# 도입한 '월'로 부터 4년 뒤 동일한 '월'에 감가상각비는 0원이 된다.
# 도입한 월보다 현재 조회하는 시기가 이전이면 감가상각비 0원부여
# 도입한 월보다 현재 조회하는 시기가 이후이면 감가상각비 현재 남은 개월 수 만큼 부여
```

현재 now함수를 사용하여 지금 시점을 고정한다.

```python
# TODO : user should put std_year and mont
std_year = 2018

now = datetime.datetime.now()
current_year = now.year 
current_month = now.month
```

만료되는 년도(expired_year)가 현재년도(2018,std_year)보다 작을 경우 0으로 depre_month을 두고,
클 경우는 모든 std_yesr을 12로 모두 동일하게 둔다.

또한, 현재월(11,current_month)이 intro_month보다 클 경우에도 0으로 depre_month를 지정한다.
작을 경우에는 12달에서 현재의 달을 빼고 이번달은 계산을 안했다고 가정해서 +1 이번달을 포함하기 위한 +1을 한다.

```python
current_month
```

```python
def get_depre_month(row):
    depre_month = 0
    if current_year < std_year:
        if row['expired_year'] < std_year:
            depre_month = 0
        elif row['expired_year'] > std_year:
            depre_month = 12
        elif row['expired_year'] == std_year:
            depre_month = row['intro_month'] - 1 
    elif current_year == std_year:
        if row['expired_year'] < std_year:
            depre_month = 0
        elif row['expired_year'] > std_year:
            depre_month = 12 - current_month + 1
        elif row['expired_year'] == std_year:
            if row['intro_month'] < current_month:
                depre_month = 0
            else:
                depre_month = row['intro_month'] - current_month + 1         
    return depre_month


df_depre['depre_month'] = df_depre.apply(lambda row: get_depre_month(row), axis=1)
```

그렇게해서 구해진 모든 변수를 이용해서 감가상각비 계산하는 함수이다.

```python
df_depre['total_cost'] = df_depre.apply(lambda row: row['monthly_cost'] * row['depre_month'], axis=1)
```
