```python
df_depre = df.iloc[:, [8, 9, 10, -2]]
df_depre.head()
```

```python
df_depre.info()
```

```python
df_depre.intro_month = df_depre.intro_month.apply(lambda month: int(month[-3:-1]))
df_depre['intro_day'] = df_depre.intro_date.apply(lambda day: int(day[-2:]))
```

```python
df_depre.head()
```

```python
def reset_intro_date(row):
    intro_date = row.intro_date
    row['intro_year'] = int(intro_date[:4])
    row['intro_month'] = int(intro_date[5:7]) 
    row['intro_day'] = int(intro_date[-2:])
    return row

df_depre = df_depre.apply(lambda row: reset_intro_date(row), axis=1)
```

```python
df_depre['monthly_cost'] = df_depre['original_price'] / 48
```

```python
df_depre.loc[df_depre.intro_month == 10]
```

```python
def get_left_month(row):
    left_month = 48
    left_month -= (std_year - row['intro_year']) * 12
    # left_month -= (std_year - row['intro_year']) * 12
    left_month -= + (std_month - row['intro_month'])
    row['left_month'] = left_month
    return row

df_depre = df_depre.apply(lambda row: get_left_month(row), axis=1)
```

```python
df_depre.loc[df_depre.intro_month==2]
```

```python
df_depre['total_cost'] = df_depre.apply(lambda row: row['monthly_cost'] * row['left_month'] if row['left_month'] < 12 else row['monthly_cost'] * 12, axis=1)
```
