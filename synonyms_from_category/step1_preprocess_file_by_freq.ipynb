```python
import pandas as pd
from itertools import chain
from elasticsearch import Elasticsearch
from elasticsearch.client import IndicesClient
from collections import Counter
```

```python
data_dir = '../synonyms_from_category/data/'
dt = '20190130'
```

```python
synonyms_dict = {}


def parse_synonym_dict(file_name):
    read_file = open(data_dir + file_name, 'r+')
    
    while True:
        line = read_file.readline()
        if not line:
            break

        info = line.split("\t")
        k = info[1]
        v = info[2].split(",")
        
        synonyms_dict[k] = v

    read_file.close()
```

```python
parse_synonym_dict('thesaurusDic_20190123.txt')
```

```python
df = pd.read_pickle(data_dir+'df_raw_190117.pkl')
```

```python
df.info()
```

```python
terms_list = df.term_vectors.tolist()
```

```python
terms_list = [term for term in chain(*terms_list)]
```

```python
word_counter = Counter(terms_list)
```

```python
word_counter = Counter(terms_list)
```

```python
df_cnt = pd.DataFrame.from_dict({tup[0]: tup[1] for tup in word_counter.most_common(len(word_counter))}, orient='index')
```

```python
df_cnt = df_cnt.reset_index()
df_cnt.columns = ['k', 'v']
```

```python
df_cnt['len_k'] = df_cnt['k'].apply(lambda k: len(k))
df_cnt
```

```python
df_temp = df_cnt.loc[(df_cnt.len_k > 1) & (df_cnt.v > 100)]
```

```python
def make_key_list(cate):
    key_list = []
    for k, v in synonyms_dict.items():
        if cate == k or cate in v:
            key_list.append(k)
    if key_list:
        return key_list
    else:
        None
```

```python
df_temp['key_list'] = df_temp['k'].apply(lambda k: make_key_list(k))
```

```python
df_temp
```

```python
key_list = df_temp.dropna(subset=['key_list'])['key_list'].tolist()
```

```python
key_list = [key for key in chain(*key_list)]
key_list = list(set(key_list))
```

```python
cand_synonym_dict = {}
for k in key_list:
    temp_list = [x for x in synonyms_dict[k] if not x.endswith('Ïö©') and not x.endswith('Ìòï') and not x.endswith('Ïãù')]
    cand_synonym_dict[k] = " ".join(temp_list)
```

```python
cand_synonym_dict
```

```python
df_synonyms = pd.DataFrame.from_dict(cand_synonym_dict, orient='index')
```

```python
df_synonyms = df_synonyms.reset_index()
df_synonyms.columns = ['k' ,'v']
df_synonyms.head()
```

```python
df_candidate = df_synonyms
# df_candidate.v = df_candidate.v.apply(lambda s: s if not str.isalpha(s) else "")
df_candidate = df_candidate.loc[(df_candidate.v != "" ) ]
df_candidate
```

```python
df_candidate.sort_values(by=['k']).to_csv(data_dir+'candidate_synonyms_'+dt+'.tsv', sep='\t', index=False)
```
