```python
import pandas as pd
import re
```

```python
dt = '190117'
data_dir = '../prd_review_v2/data/'
```

```python
df_text = pd.read_pickle(data_dir+'df_temp_text_'+dt+'.pkl')
df_term = pd.read_pickle(data_dir+'df_temp_term_'+dt+'.pkl')
```

```python
df = df_text.merge(on=['m_id'], right=df_term)
df.term_vectors = df.term_vectors.apply(lambda term_vector: term_vector.split() if term_vector else None)
df.head()
```

```python
def cleansing_text(text):
    text = re.sub(pattern='\n|ㆍ|\r|\.|,|;|:|<|>|-|_|=|\+|/|\?|!|@|\(|\)|\*|&|！|\^|#|%|\$|~|？|\[|\]|\"|[0-9]', repl=" ", string=text)
    text = re.sub(pattern='( )+', repl=' ', string=text)
    text = re.sub(pattern='(ㅎ)+', repl='ㅎ', string=text)
    text = re.sub(pattern='(ㅋ)+', repl='ㅋ', string=text)
    text = re.sub(pattern='(ㅠ|ㅜ)+', repl='ㅠ', string=text)
    return text

df.text = df.text.apply(lambda x: cleansing_text(text=x) if x else None)
df.head()
```

```python
df.sample(20)
```

```python
df = df.dropna(subset=['text', 'term_vectors'])
```

```python
df = df.reset_index(drop=True)
df.to_pickle(data_dir+'df_raw_'+dt+'.pkl')
```
