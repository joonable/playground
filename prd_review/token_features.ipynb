```python
import pandas as pd
import re
import numpy as np
from itertools import chain
from collections import Counter
```

```python
df = pd.read_pickle('./prd_data.pkl')
df.head()
```

```python
df = df.dropna(subset=['text']).reset_index(drop=True)
```

```python
corpus = " ".join(df.text.tolist())
```

```python
corpus = re.sub(pattern='\n|ㆍ|\r|\.|,|;|:|<|>|-|_|=|\+|/|\?|!|@|\(|\)|\*|&|！|\^|#|%|\$|~|？|\[|\]|\"|[0-9]', repl=" ", string=corpus)
corpus = re.sub(pattern='( )+', repl=' ', string=corpus)
corpus = re.sub(pattern='(ㅎ)+', repl='ㅎ', string=corpus)
corpus = re.sub(pattern='(ㅋ)+', repl='ㅋ', string=corpus)
```

```python
corpus_list = corpus.split(" ")
```

```python
subfix = list('을를이가는은에도')
token_list = []
for token in corpus_list :
    if token[-1] in subfix and len(token) > 2:
        token_list.append(token[:-1])
    else:
        token_list.append(token)
        
token_list = [token for token in token_list if token[-1] not in ['요', '다']]
```

```python
token_count = Counter(token_list)
token_count.most_common(10)
```

```python
terms_list = []
for x in df.sorted_terms:
    if x:
        terms_list.extend(x)
```

```python
terms_count = Counter(terms_list)
terms_count.most_common(10)
```

```python
tokens = set([tup[0] for tup in token_count.most_common(5000)])
terms = set([k for k, v in terms_count.items()])
```
